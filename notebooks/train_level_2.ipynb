```python
import plotly.graph_objs as go
from copy import deepcopy
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
from json import dumps
```

```python
df = pd.read_csv('./dataset/train.csv',index_col='Id')
df
```

```python
col_catagorial = ['Neighborhood', 'Exterior2nd', 'MSSubClass','Exterior1st', 'YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'MoSold','TotRmsAbvGrd', 'OverallQual', 'OverallCond', 'Condition1', 'SaleType','RoofMatl', 'BedroomAbvGr', 'Condition2', 'HouseStyle', 'Functional','GarageType', 'SaleCondition', 'RoofStyle', 'Electrical', 'Heating','BsmtFinType2', 'BsmtFinType1', 'Foundation', 'BldgType', 'MSZoning','GarageCond', 'YrSold', 'GarageQual', 'ExterCond', 'GarageCars','LotConfig', 'HeatingQC', 'Fireplaces', 'LandContour', 'KitchenQual','MasVnrType', 'FullBath', 'BsmtFullBath', 'KitchenAbvGr', 'LotShape','BsmtExposure', 'BsmtCond', 'BsmtQual', 'ExterQual', 'HalfBath','GarageFinish', 'BsmtHalfBath', 'LandSlope', 'PavedDrive', 'Utilities','Street', 'CentralAir','FireplaceQu','SalePrice']
df = deepcopy(df[col_catagorial])
df
```

```python
for col in df.columns:
    print(col)
    display(df[col].value_counts())
```

```python
features = df.columns.to_list()
features.remove('SalePrice')
print(features)
```

```python
# most important dictionary.
data_dict_median = {}
```

```python
for col in features:
    group_by_catagory = df[[col,'SalePrice']].groupby(col)['SalePrice'].median()
    group_by_catagory.loc['Other'] = group_by_catagory.mean()
    data_dict_median[col] = group_by_catagory.to_dict()
```

```python
data_dict_median
```

```python
with open('./data_dict_median.json','wt') as f:
    f.write(dumps(data_dict_median))
```
