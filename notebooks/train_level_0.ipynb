```python
import plotly.graph_objs as go
from copy import deepcopy
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv('./FilteredCsv/dataset_after_level_1.csv',index_col='Id')
df
```

```python
col_continuous = ['GrLivArea','LotArea', 'BsmtFinSF1', 'BsmtUnfSF', '1stFlrSF', 'TotalBsmtSF', 'GarageArea', '2ndFlrSF', 'WoodDeckSF', 'OpenPorchSF','BsmtFinSF2', 'EnclosedPorch', 'LotFrontage','ScreenPorch','SalePrice']
df = deepcopy(df[col_continuous])
```

```python
df
```

```python
df.isna().sum()
```

```python
df['LotFrontage'].fillna(df['LotFrontage'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
features = df.drop(columns='SalePrice').to_numpy()
features
```

```python
target = df['SalePrice'].to_numpy().reshape(-1,1)
target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(features,target,random_state=2,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train.flatten())
rfr.score(x_test,y_test)
```

```python
y_pred_lr = lr.predict(x_test)
y_pred_rfr = rfr.predict(x_test)
```

```python
fig_score = go.Figure()
fig_score.add_scatter(x=y_test.flatten(),y=y_pred_lr.flatten(),mode='markers')
fig_score.layout.xaxis.title = 'Actual SalePrice $'
fig_score.layout.yaxis.title = 'Predicted SalePrice $'
fig_score.show()
```

```python
df_score  = pd.DataFrame({'y_test':y_test.flatten(),'y_pred_lr':y_pred_lr.flatten(),'y_pred_rfr':y_pred_rfr.flatten()})
df_corpse = df_score.sort_values(by='y_test').reset_index(drop=True)
```

```python
df_corpse
```

```python
fig_score = go.Figure()
fig_score.add_scatter(x=df_corpse.index,y=df_corpse['y_test'],mode='lines',name='Actual SalePrice')
fig_score.add_scatter(x=df_corpse.index,y=df_corpse['y_pred_lr'],mode='markers',name='Prediction_lr')
fig_score.add_scatter(x=df_corpse.index,y=df_corpse['y_pred_rfr'],mode='markers',name='Prediction_rfr')

fig_score.update_traces({'marker':{'size':5}})
fig_score.layout.yaxis.title = 'Sale_Price'
fig_score.show()
```
